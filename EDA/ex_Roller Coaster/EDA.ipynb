### Step 2: Data Preperation

* Dropping Irrelevant columns and rows

```python
df.columns
```

```python
feature = ['coaster_name', 'Manufacturer', 'Type_Main', 'Location', 'Status','Opened', 'year_introduced', 'latitude', 'longitude', 
       'opening_date_clean',  'speed_mph',  'height_unit', 'height_ft',
       'Inversions_clean', 'Gforce_clean']

df_ = df[feature].copy()
```

```python
#changing date column to datetime type 
df_['opening_date_clean'] = pd.to_datetime(df_['opening_date_clean'])
```

```python
# Rename columns

df_ = df_.rename(columns={'coaster_name': 'Coaster_Name',
                   'year_introduced': 'Year_Introduced'
                   })
```

```python
#check for duplicated rows
df_.duplicated()
```

```python
#checking duplicate
df_.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
#check for subset of columns
# adding  '~' means NOT
df_ = df_.loc[~df.duplicated(subset=['coaster_name','Location', 'Opening date'])] \
.reset_index(drop = True)  # reset the index becuz come rows were dropped
```

# Step 5 : Ask a Question about the data

* What are the locations with the fastest roller coasters? (min of 10)

```python
df_grp = df_.groupby('Location')['speed_mph'].agg(['count','mean'])
df_grp.query('count >= 10 and Location != "Other"').sort_values(by='count')
```

```python
ax = df.query('Location != "Other"')\
    .groupby('Location')['speed_mph']\
    .agg(['mean','count'])\
    .query('count >= 10')\
    .sort_values('mean')['mean']\
    .plot(kind='barh', figsize=(12,5), title='Average Coast Speed')

ax.set_xlabel('Average Coaster Speed')
plt.show()
```
